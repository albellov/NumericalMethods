## \begin{equation} \nonumber
    (1-\lambda)^{2m} = 10^{-2m}\\
    1-\lambda = \sqrt[2m]{10^{-2m}}\\
    \lambda = 1 - \sqrt[2m]{10^{-2m}}\\
    \lambda = 1 - 0.1 e^{\frac{2\pi k}{2m}}; k = 0,1,...,2m-1
\end{equation}

```python
import numpy as np

def my_matrix(m):
    
    A = np.eye(2*m)
    A[0,2*m-1] = 10**(-2*m)
    A = A + np.diag(np.ones(2*m-1),-1)
    
    values,_ = np.linalg.eig(A)

    return values
        
    
def draw_complex_roots(compl):
    
    plt.figure(figsize=(10,8))
    
    c_real = [c.real for c in compl ]
    c_imag = [c.imag for c in compl ]
    
    plt.plot(c_real,c_imag,'b*',markersize = 12)
        
    plt.grid()
    plt.xlabel('Real',fontsize=16)
    plt.ylabel('Imag',fontsize=16)
    plt.title('Eigenvalue of the matrix',fontsize=15)
    plt.show()   
    
eig = my_matrix(20)
draw_complex_roots(eig)
```

## №4
Вычислить интеграл 
### \begin{equation} \nonumber
	\int_{0}^{1} \frac{4}{1 + x^2} dx
\end{equation}
численно по формуле трапеций при числе интервалов сетки $N = 16,32,64,128,256$. Затем с помощью формулы __(1)__ или формулы __(2)__ построить график эффективного порядка. Провести уточнение решения по формуле Ричардсона, получив из 5 расчетных значений 4 уточненных. По уточненным значениям вновь построить график эффективного порядка.

### \begin{equation} \label{c1eq13}
	p = \log_r \frac{U_{r^2h} - U_{rh}}{U_{rh} - U_h} (1)
\end{equation}
### \begin{equation} \label{c1eq14}
	\lg \left| U_{rh} - U_h \right| = \lg \big( \left| c \right| \left( 1 - r^p \right) \big) + \left( -p \right) \left( - \lg h \right) (2)
\end{equation}

```python
import traceback
import numpy as np

def my_fun_4(x):
    return 4/(1+x**2)
def my_fun_5(x):
    return 1/2/np.sqrt(x)

def Trapezoidal_method(fun,a,b,N):
    try:
        
        if N%2 == 1:
            raise ValueError(a)
        h = (b-a)/N
        values_into_range = np.array([fun(a+h*x_j) for x_j in range(1,N)])
        
        int_sum = h*( (fun(a)+fun(b))/2 + values_into_range.sum())
        return int_sum
        
    except ValueError:
        print('Enter an even N\n',traceback.format_exc())
        
def Middle_sum(fun,a,b,N):
    try:
        
        if N%2 == 1:
            raise ValueError(a)
        h = (b-a)/N
        values_into_range = np.array([fun(a+h*i - h/2) for i in range(1,N+1)])
        
        int_sum = h*values_into_range.sum()
        return int_sum
        
    except ValueError:
        print('Enter an even N\n',traceback.format_exc())
```

```python
traperz_values = [Trapezoidal_method(my_fun_4,0,1,2**i) for i in range(1,9)]

traperz_values
```

```python
import matplotlib.pyplot as plt

def effective_order_of_method(U,r,
                              lable = 'График эффективного порядка',
                              draw_flag = False):
    try:
        
        if len(U) < 3:
            raise IndexError()
      #  exp_p = [ (U[i]-U[i+1])/(U[i+1]-U[i+2]) for i in range(len(U)-2) ]
        exp_p = [ (U[i+2]-U[i+1])/(U[i+1]-U[i]) for i in range(len(U)-2) ]
        p = np.log(exp_p)/np.log(r)
        

        if draw_flag:
            plt.figure(figsize=(10,8))
            plt.plot(range(1,len(p)+1), p,'b*-')    
            plt.grid()
            plt.xlabel('№ triple',fontsize=16)
            plt.ylabel(r'$p$',fontsize=16)
            plt.title(lable,fontsize=15)
            plt.show()
            
        return p
        
    except IndexError:
        print('More than three values are required\n',traceback.format_exc())
        return None
 
def Richardsons_clarification(method, p):

    r = 1/2
    N = [2**i for i in range(1,9)]
    U_total = []
    
    for ind in range(len(N)-1):
        
        U_h = method( my_fun_4,0,1, N[ind] )
        U_rh = method( my_fun_4,0,1, N[ind+1] )
        
        U_total.append( ( r**(-p)*U_rh - U_h )/( r**(-p) - 1 ) )
    
    return U_total[1:]
```

```python
values_eff = effective_order_of_method(traperz_values[:6],1/2.,draw_flag=True)
```

```python
rich_clar = Richardsons_clarification(Trapezoidal_method, 2)   
values_eff = effective_order_of_method(rich_clar,1/2,'График эффективного порядка с уточнением Ричардсона',True)
```
